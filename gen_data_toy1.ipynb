```python
from drawdata import ScatterWidget
```

```python
widget = ScatterWidget()
widget
```

```python
df = widget.data_as_pandas[["x","y","label"]]
df.head()
```

```python
df.columns = ["x1","x2","y"]
```

```python
df.head()
```

```python
df_y = df.groupby("y")["y"].count()
df_y
```

```python
ax = df_y.plot(kind="bar", title="Numero de valores por clases")
ax.set_xlabel("Clase")
ax.set_ylabel("Numero de valores")
ax.set_xticklabels(ax.get_xticklabels(), rotation=0)
ax.grid(ls=":")
```

```python
label2num = {'a':0, 'b':1, 'c':2, 'd':3}
num2label = {0:'a', 1:'b', 2:'c', 3:'d'}
```

```python
df["y"] = df["y"].map(label2num)
df.head()
```

```python
import seaborn as sb

sb.scatterplot(data=df, x="x1", y="x2", hue="y", palette="Set1")
```

```python
df.to_csv("data.csv", index=False)
```
